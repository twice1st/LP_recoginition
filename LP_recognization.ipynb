```python
from sklearn.model_selection import train_test_split
import os
import shutil
```

```python
def split_data():
  path = '/content/GreenParking_Recog/images'
  lst_name_data = os.listdir(path)
  X, X_test = train_test_split(lst_name_data, random_state=42, test_size=0.1)
  X_train, X_valid = train_test_split(X, random_state=42, test_size=0.1)

  return X_train, X_valid, X_test
```

```python
def copy_data(X, cas):
  path_img = '/content/GreenParking_Recog/images/'
  path_lab = '/content/GreenParking_Recog/labels/'
  dst = '/content/GreenParking_Recog/' + cas
  for name in X:
    shutil.copy(path_img + name, dst + '/images')
    shutil.copy(path_lab + name.split('.')[0] + '.txt', dst + '/labels')
```

```python
X_train, X_valid, X_test = split_data()
```

```python
copy_data(X_train, 'train')
copy_data(X_test, 'test')
copy_data(X_valid, 'valid')
```
